```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# veriyi okuma işlemi
df = pd.read_csv("AB_US_2020.csv")
```

```python
df.head(10)
```

```python
df.describe().T
```

```python
# boş değerler
df.isnull().sum()
```

```python
df.info()
```

```python
# astype ile conver et object to category
df["room_type"] = df["room_type"].astype("category")
```

```python
df["room_type"].info()
```

```python
# nan değerleri doldurma işlemi
df = df.fillna(0)
```

```python
df.isnull().sum()
```

```python
# oda tipine göre fiyat dağılışını inceleyelim
import matplotlib.pyplot as plt
sns.boxplot(x = "room_type",y = "price",data=df, color="skyblue");
plt.show()
```

```python
# oda tipine göre fiyat dağılışını inceleyelim
sns.catplot(x = "room_type",y = "price",data=df);
```

```python
# oda tipine göre fiyat dağılışını inceleyelim
sns.barplot(x = "room_type",y = "price",data=df);
```

```python
# fiyatların gecelere göre dağılımı
sns.barplot(x = "longitude",y = "price",data=df);
plt.show()
```

```python
# oda tipine göre fiyat dağılışını inceleyelim
sns.catplot(x = "room_type",y = "price", kind="violin", data=df);
```

```python
# fiyatların şehirlere göre dağılımını inceleyelim
sns.barplot(x = "city",y = "price",data=df);
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# fiyat eğrisi
sns.catplot(y = "price", kind="violin", data=df);
```

```python
# şehirlerin değerlendirilmesi
sns.catplot(x = "city",y = "number_of_reviews", kind="violin", data=df);
plt.xticks(rotation=45, ha="right")
```

```python
sns.catplot(x = "room_type",y = "price", kind="violin", data=df);
plt.xticks(rotation=45, ha="right")
```
