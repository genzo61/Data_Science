```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# veriyi okuma i≈ülemi
df = pd.read_csv("AB_US_2020.csv")
```

```python
# astype ile conver et object to category
df["room_type"] = df["room_type"].astype("category")
```

```python
df["room_type"].info()
```

```python
# nan deƒüerleri doldurma i≈ülemi
df = df.fillna(0)
```

```python
df.isnull().sum()
```

```python
# oda tipine g√∂re fiyat daƒüƒ±lƒ±≈üƒ±nƒ± inceleyelim
import matplotlib.pyplot as plt
sns.boxplot(x = "room_type",y = "price",data=df, color="skyblue");
plt.show()
```

```python
# oda tipine g√∂re fiyat daƒüƒ±lƒ±≈üƒ±nƒ± inceleyelim
sns.catplot(x = "room_type",y = "price",data=df);
```

```python
# oda tipine g√∂re fiyat daƒüƒ±lƒ±≈üƒ±nƒ± inceleyelim
sns.barplot(x = "room_type",y = "price",data=df);
```

```python
# fiyatlarƒ±n gecelere g√∂re daƒüƒ±lƒ±mƒ±
sns.barplot(x = "longitude",y = "price",data=df);
plt.show()
```

```python
# oda tipine g√∂re fiyat daƒüƒ±lƒ±≈üƒ±nƒ± inceleyelim
sns.catplot(x = "room_type",y = "price", kind="violin", data=df);
```

```python
# fiyatlarƒ±n ≈üehirlere g√∂re daƒüƒ±lƒ±mƒ±nƒ± inceleyelim
sns.barplot(x = "city",y = "price",data=df);
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# fiyat eƒürisi
sns.catplot(y = "price", kind="violin", data=df);
```

```python
# ≈üehirlerin deƒüerlendirilmesi
sns.catplot(x = "city",y = "number_of_reviews", kind="violin", data=df);
plt.xticks(rotation=45, ha="right")
```
